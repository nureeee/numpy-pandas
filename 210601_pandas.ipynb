### 시리즈가 여러개 모여서 하나의 데이터 시트를 표현하면 데이터프레임이 된다.

```python
sample_data = [['이누리', 22.9, 159],  ## 1번 방법
               ['이리릴', 20, 155],
               ['이로로', 23, 156]]

df1 = pd.DataFrame(sample_data)
```

```python
df1.columns = ['이름','나이','키']
df1
```

```python
##방법_2 dict로 만들기
sample_data = {
    '이름':['이', '김', '박'],
    '나이':[22, 24, 23],
    '키':[156, 158, 159]
}
df2 = pd.DataFrame(sample_data)
```

```python
##방법_2 dict로 만들기
sample_data = {
    '이름':['이', '김', '박'],
    '나이':[22, 24, 23, 26],  ##데이터의 수가 맞지 않아 에러가 발생한다.
    '키':[156, 158, 159]
}
df2 = pd.DataFrame(sample_data)
```
